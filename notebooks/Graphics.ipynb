# 1. $f$-divergences

```julia
f1(s) = s.*log(s) - s +1
f2(s) = abs(s-1)
xmax, ymax = 4.5, 4.
X = linspace(1e-20,xmax,300)
fig, ax = subplots(figsize=[5,3],1,1)
plot(X,f1(X),"k");
plot(X,f2(X),"-.k");
plot([0.4, 0.4, 2.5, 2.5], [ymax+1, 0, 0, ymax+1], ".--k")
plot([1., 1., 2.5, 2.5], [ymax+1, 0, 0, ymax+1], ":k")

legend([L"KL",L"TV", L"RG", L"\iota_{=}"],4)
axis([0 ,xmax, -0.5, ymax])
yax = ax[:axes][:get_yaxis]();
yax[:set_visible](true)
ax[:set_frame_on](false)
xax= ax[:get_xaxis]()
xax[:tick_bottom]()
yax[:tick_left]()
xticks([0,1,2,3,4]),yticks([0,1,2,3]);
tight_layout()
#savefig("outputs/fdivergences.pdf")
```

# 2. $Prox^{KL}_f$

Plot $r = f\left(\frac{d\mu}{d\nu}\right)$

```julia
ϵ, λ1,  λ2 = 1.3, .5, .8
c, d = 0.5, 3.5
#xmax = 2.5; ymax = 4.
f(s,λ) = (1./s).^(λ/(λ+ϵ))
g(s,λ) =  max(exp(-λ/ϵ),min(exp(λ/ϵ),1./s))
h(s,λ) =  max(c./s, min(d./s,1))
X = linspace(0,20,300)
fig, ax = subplots(figsize=[5,3],1,1)
loglog(X,f(X,3.),"k")
loglog(X,g(X,λ1),"-.k")
loglog(X,h(X,λ1),"--k")
#plot(X,f(X,λ2),"-.k")
#plot(X,g(X,λ2),"-.r")
#plot(X,f(X,λ1),"k")
loglog(X,1./X,":k")
axis([.05 ,20., .2, 4.])
yax = ax[:axes][:get_yaxis]();
yax[:set_visible](true)
ax[:set_frame_on](false)
xax= ax[:get_xaxis]()
xax[:tick_bottom]()
yax[:tick_left]()
#legend([L"\psi",L"f_{TV}",L"f_{th}"],"east")
#ylabel(L"prox^{K L}_{D_{\varphi}(\cdot |1)}(s)")
ylabel(L"proxdiv$_F(s)$")
xlabel(L"s")
#xticks([0,1,2]),yticks([0,1,2,3]);
tight_layout()
#savefig("outputs/proxs.pdf")
```

```julia
ϵ, λ1,  λ2 = 1., .5, .8
c, d = 0.5, 3.5
xmax = 2.5; ymax = 3.
f(s,λ) = s.^(λ/(λ+ϵ))
g(s,λ) =  max(exp(-λ/ϵ),min(exp(λ/ϵ),s))
h(s,λ) =  max(s.*c, min(s.*d,1))
Y = logspace(-2,2,300)
fig, ax = subplots(figsize=[5,3],1,1)
semilogx(Y,log(f(1./Y,λ1).*Y),"k")
semilogx(Y,log(g(1./Y,λ1).*Y),"-.k")
semilogx(Y,log(h(1./Y,λ1).*Y),"--k")
#plot(X,f(X,λ2),"-.k")
#plot(X,g(X,λ2),"-.r")
#plot(X,f(X,λ1),"k")
semilogx(Y,log(Y./Y),":k")
#axis([0 ,xmax, 0, ymax])
yax = ax[:axes][:get_yaxis]();
yax[:set_visible](true)
ax[:set_frame_on](false)
xax= ax[:get_xaxis]()
xax[:tick_bottom]()
yax[:tick_left]()
#legend([L"\psi",L"f_{TV}",L"f_{th}"],"east")
ylabel(L"\log(prox^{KL}_{D_f(\cdot |\mu)}(\nu))")
xlabel(L"\nu")
#xticks([0,1,2]),yticks([0,1,2,3]);
tight_layout()
```

```julia
X = linspace(0,3,1000)
c = .8
f1(x)= x.*log(x)-x+1
f2(x)= x.*log(x/exp(-c))-x+exp(-c)
fig, ax = subplots(figsize=[5,3],1,1)
plot(X,f1(X),"k")
plot(X,c.*X,"--k")
plot(X,c.*X+f1(X),":k")
plot(X,f2(X),"k",linewidth=2)
yax = ax[:axes][:get_yaxis]();
yax[:set_visible](true)
ax[:set_frame_on](false)
xax= ax[:get_xaxis]()
xax[:tick_bottom]()
yax[:tick_left]()
legend([L"\varepsilon\,\mathrm{KL}(x|1)",L"c\cdot x",L"(1)+(2)", 
    L"\varepsilon\,\mathrm{KL}(x|e^{-c/\varepsilon})"],"east")
ylabel(L"f(x)")
xlabel(L"x")
xticks([0,1,2,3]),yticks([0,1,2,3]);
tight_layout()
#savefig("outputs/KLabsorb.pdf")
```
