### Gradient flow of TV in 1D

```julia
# THE NUMERICAL BALL FOR BUILDING INITIAL DENSITIES
"ball B(c,r) in [0,1]^d"
function ball(c,r;N=50) 
    X = linspace(.5/N,1-.5/N,N)
    if length(c) == 1
        return [ (abs(x-c[1])<=r)*1. for x in X]
    elseif length(c) == 2
        return [ ((x-c[1])^2+(y-c[2])^2<r^2)*1. for x in X, y in X]
    end
end
```

```julia
dim, N = 1, 600 #700
X = linspace(.5/N,1-.5/N,N)
p = ball(.5,.2,N=N)+ball(.35,.05,N=N)+2*ball(.65,.05,N=N)#+ball(.35,.05,N=N)
p[1]=0;p[end]=0
figure(figsize=[6,3])
fill(X,p,color=[.5,.5,.8]); title("Initial density")
```

```julia
nstep, ϵ = 20, 1e-6 #2e-6
tauvec = logspace(-2.7,0,nstep)
flow, u, v, w = fill(p,nstep+1),fill(zeros(p),nstep+1),fill(zeros(p),nstep+1),fill(zeros(p),nstep+1)
for k=1:nstep
    println("Step number $(k).")#600 niter
    @time flow[k+1],u[k+1],v[k+1],w[k+1] = gradientflowTV.onestepflow(flow[k], tauvec[k], dim, ϵ;niter=600,epsstep=8., tol=0.0);
end
```

```julia
figure(figsize=[15,4])
mm=length(flow)
ind = cat(1,1:1:length(flow))
subplot(131)
for l = ind
    plot(X,flow[l],color=[l/mm; .5; .5])
    axis([0;1;0;3.2]); #title("Flow")
end

subplot(132)
for l = setdiff(ind,1)
    plot(X,u[l],color=[l/mm; .5; .5])
    axis([0;1;-1*1e-3;3*1e-3]); #title("Dual variable u")
end

subplot(133)
for l = setdiff(ind,1)
    plot(X,v[l]+w[l],color=[l/mm; .5; .5])
    axis([0;1;-.001;.003]); #title("Dual variable v+w")
end

#savefig("outputs/GFTV_5.pdf");
```

## 2 dimensional: in progress

In 2d, we use the following anisotropic discretization of TV for each quadruplet of points 
$$f(a,b,c,d)=\left( (a-b)^2 + (c-d)^2 + (a-c)^2 + (b-d)^2 \right)^\tfrac12$$
This implies that each proximal step requires to find the minimizer of $g = \epsilon KL + f$.
